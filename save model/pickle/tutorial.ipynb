```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
from sklearn import linear_model;
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
plt.scatter(df.area, df.price, color='red', marker='+')
plt.xlabel('Area (sqft)')
plt.ylabel('Price (USD)')
```

```python
l_g = linear_model.LinearRegression()
l_g.fit(df[['area']], df.price)
```

```python
predictval = l_g.predict([[3300]])
print(predictval)
```

```python
l_g.coef_, l_g.intercept_ #m, c
```

SAVING MODEL USING PICKLE

```python
import pickle
```

```python
with open ('model_pickle', 'wb') as f:
    pickle.dump(l_g, f)
```

```python
with open ('model_pickle', 'rb') as f:
    model = pickle.load(f)
```

```python
model.predict([[5000]])
```
