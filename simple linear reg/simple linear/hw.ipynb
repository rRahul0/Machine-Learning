```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
from sklearn import linear_model;
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
plt.scatter(df.year, df.income, color='red', marker='+')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
```

```python
l_reg = linear_model.LinearRegression()
l_reg.fit(df[['year']], df.income)
```

```python
l_reg.predict([[2024]])
```

```python
plt.scatter(df.year, df.income, color='red', marker='.')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.plot(df.year, l_reg.predict(df[['year']]), color='blue')
```

```python
l_reg.score(df[['year']], df.income)*100
```
