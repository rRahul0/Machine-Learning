```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
from sklearn import linear_model;
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
plt.scatter(df.area, df.price, color='red', marker='+')
plt.xlabel('Area (sqft)')
plt.ylabel('Price (USD)')
```

```python
l_g = linear_model.LinearRegression()
l_g.fit(df[['area']], df.price)
```

```python
predictval = l_g.predict([[3300]])
print(predictval)
```

```python
l_g.coef_, l_g.intercept_ #m, c
```

```python
d=pd.read_csv('areas.csv')
```

```python
p=l_g.predict(d[['area']])
d['price']=p
```

```python
d.to_csv('predicted_home_prices.csv', index=False)
```
